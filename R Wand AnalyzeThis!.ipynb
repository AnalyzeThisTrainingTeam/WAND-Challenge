```r
# use ridge regression
library(MASS)
# Need to convert from list/vector to "values"
str(y <- unlist(good_data["NetSales"]))
model.ridge <- lm.ridge(y ~ ., good_data[my_cols], lambda = seq(0,2500,0.1))
```

```r
plot(seq(0,2500,0.1), model.ridge$GCV, main="GCV of Ridge Regression", type="l", 
    xlab=expression(lambda), ylab="GCV")
```

```r
# The optimal lambda is given by
str(lambda.ridge <- seq(0,2500,0.1)[which.min(model.ridge$GCV)])
```

```r
str(beta.ridge <- coef(model.ridge)[lambda.ridge,])
resid.ridge <- y - beta.ridge[1] - as.matrix(good_data[,my_cols])%*%beta.ridge[2:68]

# To find estimated degrees of freedom
d <- svd(as.matrix(good_data[,my_cols]))$d
str(df <- nrow(good_data) - sum(d^2/(lambda.ridge+d^2)))

str(rss.ridge <- sum(resid.ridge^2)/df)
```
